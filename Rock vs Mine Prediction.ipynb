```python
#Importing the dependencies
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
#Loding the csv file
sonar_data=pd.read_csv("sonar data.csv",header=None)
sonar_data
```

```python
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
# Find the number of rocks and  mines in dataset
sonar_data[60].value_counts()
```

```python
# Based on the last column grouup by the rest of the column and find the mean values
sonar_data.groupby(by=60).mean()
```

```python
#Seperating the input feature and target column
#Seperating the input feature and target column
X = sonar_data.drop(columns=[60], axis=1)
Y = sonar_data[60]
```

```python
print(X)
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Model Training
model=LogisticRegression()
```

```python
#Training the logistics regression model with training data
model.fit(X_train,Y_train)
```

```python
#Model Evaluation
#Accuracy score for training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print("Accuracy score of training data",training_data_accuracy)
```

```python
#Model Evaluation
#Accuracy score for testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print("Accuracy score of training data",testing_data_accuracy)
```
